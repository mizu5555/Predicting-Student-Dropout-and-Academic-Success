```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
df = pd.read_csv('data.csv', delimiter=';')
df.head()
df.rename(columns={'Nacionality':'Nationality','Daytime/evening attendance\t':'Daytime/evening attendance'}, inplace=True)
df = df[df['Target'] != 'Enrolled']
df
```

```python
# svm part import
from sklearn import svm
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
# svm all dataset
svm_x = df.drop(columns=['Target'])
svm_y = df['Target']

svm_data = []

for i in range(5):
    scaler = StandardScaler()
    svmX_train, svmX_test, svmY_train, svmY_test = train_test_split(svm_x, svm_y, test_size=0.2, random_state=i)
    svmX_train = scaler.fit_transform(svmX_train)
    svmX_test = scaler.transform(svmX_test)
    svm_model = svm.SVC(kernel='rbf')
    svm_data.append({'svm_model': svm_model, 'train_data': svmX_train, 'train_labels': svmY_train, 'test_data': svmX_test, 'test_labels': svmY_test})
```

```python
# svm all run
meanAcc = 0.0

for i in range(5):
    svm_data[i]['svm_model'].fit(svm_data[i]['train_data'], svm_data[i]['train_labels'])
    y_pred = svm_data[i]['svm_model'].predict(svm_data[i]['test_data'])
    accuracy = accuracy_score(svm_data[i]['test_labels'], y_pred)
    meanAcc = meanAcc + accuracy
    print('Test' + str(i+1) + ' Accuracy:' + str(accuracy))
    
print('Average Accuracy:' + str(meanAcc/5))
```
